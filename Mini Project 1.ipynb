```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# I expect stocks in technology and crptocurrency would have high volatility, and stocks in more traditional sectors have lower volatility.
high_risk_tickers = ['TSLA', 'NVDA', 'COIN']
low_risk_tickers = ['JNJ', 'PG', 'WMT']
tickers = high_risk_tickers + low_risk_tickers
data = yf.download(tickers, start="2024-06-01", end="2025-06-01")
```

```python
data.head(5)
```

```python
# The below results somehow verifies my intuition.

data = data['Close']

# Compute log returns
log_returns = np.log(data / data.shift(1)).dropna()

# High risk portfolio
high_risk_returns = log_returns[high_risk_tickers].mean(axis=1)
high_risk_vol = high_risk_returns.std() * np.sqrt(252)

# Low risk portfolio
low_risk_returns = log_returns[low_risk_tickers].mean(axis=1)
low_risk_vol = low_risk_returns.std() * np.sqrt(252)

# Mean returns
high_risk_mean = high_risk_returns.mean() * 252
low_risk_mean = low_risk_returns.mean() * 252

# Print results
print("High Risk Portfolio:")
print(f"  Annual Return: {high_risk_mean:.2%}")
print(f"  Annual Volatility: {high_risk_vol:.2%}")

print("\nLow Risk Portfolio:")
print(f"  Annual Return: {low_risk_mean:.2%}")
print(f"  Annual Volatility: {low_risk_vol:.2%}")
```
